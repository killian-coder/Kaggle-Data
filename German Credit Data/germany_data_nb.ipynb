### 1.0 Imports for Data Visualization

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rcParams
%matplotlib inline

# figure size in inches
rcParams['figure.figsize'] = 10,6
```

```python
data = pd.read_csv('german_credit_data_with_risk.csv')
data.drop(data.columns[0], 1, inplace = True) # Drop The custom index column
data.head()
```

```python
# LETS CHECK FOR THE NUMBER OF UNIQUE VALUES IN EACH COLUMN
print(data.nunique())
```

```python
# CHECK FOR AGE DISTRIBUTION
fig, ax = plt.subplots(2,1, figsize = (15,10))
sns.distplot(data['Age'], ax = ax[0])
sns.countplot(x = 'Age', data = data, palette = 'Set1', hue = 'Sex', ax = ax[1])
plt.show()
```

```python
# We look at credit amount column
interval = (18,25,35,60,120)
categories = ['Student', 'Young', 'Adult', 'Senior']
data['Age_cats'] = pd.cut(data.Age, interval, labels = categories)

ax = sns.factorplot(x = 'Age_cats', y = 'Credit amount', data = data, 
                    kind = 'box', size = 5, aspect = 2, hue = 'Risk', palette = 'Set1')

ax.set(xlabel='Age Categorical', ylabel='Credit Amount(U$)',
       title="Age Categorical x Credit Amount Separeted by Risk groups")

plt.show()
```

### We will now Look at the distribuition of Housing own and rent by Risk
From the plot below, we can see that the own and good risk have a high correlation

```python
ax = sns.countplot(x = 'Housing', data = data, hue = 'Risk', palette = 'Set1')
ax.set(xlabel = 'Housing Categories', ylabel = 'Counts', title = 'Housing Count')
plt.show()
```

We can see below that we have the highest values are in the free category and we get a different distribution by Risk

```python
ax = sns.factorplot(x = 'Housing', y = 'Credit amount', data = data, 
               kind = 'violin', hue = 'Risk', palette = 'Set1',size=4,aspect=2,split=True)

ax.set(xlabel = 'Housing Category', ylabel = 'Credit Amount ($)',title = 'Housing by credit amount')
plt.show()
```

```python
print('Total difference by sex: ')
print(data.groupby("Sex")['Sex'].count())
sns.countplot(x = 'Sex', data = data, hue = 'Risk', palette = 'Set1')
plt.show()
```

### Convert String data to Integers for scikit-learn:

```python
data.Purpose.replace(('radio/TV', 'education','furniture/equipment', 'car', 'business', 'domestic appliances','repairs','vacation/others'), (0,1,2,3,4,5,6,7), inplace=True)

data.Sex.replace(('female','male'), (0,1), inplace=True)

data.Housing.replace(('own','free','rent'), (0,1,2), inplace=True)

data["Saving accounts"].replace((str('nan'), 'little', 'quite rich', 'rich', 'moderate'), (0,1,3,4,2), inplace=True)

data.Risk.replace(('good', 'bad'),(0,1), inplace=True)

data["Checking account"].replace(('little', 'moderate', 'rich'), (0,1,2), inplace=True)

data["Age_cats"].replace(('Student', 'Young', 'Adult','Senior'), (0,1,2,3), inplace=True)
```
